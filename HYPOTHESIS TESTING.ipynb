```python
H0 - The sample has a Gausian distribution
H1 = The sample does not have a Gausian dist
```

```python
# Example of the shpiro-Wilk Normality test
from scipy.stats import shapiro  # used to test corelation 
data = [0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, -1.478, -1.637, -1.869]
stat, p = shapiro(data)
print( 'stat = %.3f, p=%.3f'  %(stat,p))

if p > 0.05:
    print('Probably Gausian')
else:
    print('Probably no Gausin')
```

```python
import seaborn as sns
sns.distplot(data)
```
