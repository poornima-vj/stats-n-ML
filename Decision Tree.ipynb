```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier  #import Decion tree
from sklearn.model_selection import train_test_split
from sklearn import metrics #for accuaracy calc
from sklearn.metrics import classification_report
```

```python
df.columns
```

```python
#split dataset in features and target variable
# feature_columns = ['preg', 'plas', 'pres', 'skin', 'insu', 'mass', 'pedi', 'age']
# target_column = ['class']

X = df.drop(['class'] , axis=1)
y = df['class']
```

```python
y.value_counts()
```

```python
X.describe()
```

```python
# split the dataset into training and test set

X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size = 0.3,
                                                    stratify = None)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
# create Decision tree classifier object

clf = DecisionTreeClassifier()
```

```python
# Train Decision tree classifier object

clf = clf.fit(X_train, y_train)
```

```python
# Predict ths response for test dataset
y_pred = clf.predict(X_test)
```

```python
# Model Accuaracy, how often is the classifier
print("test Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
y_pred_train = clf.predict(X_train)
```

```python
print("test Accuracy:", metrics.accuracy_score(y_pred_train, y_train))
```

```python
print(classification_report(y_test, y_pred))
```
