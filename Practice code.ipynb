```python
import pandas as pd
import numpy as np
import datetime
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/manishanker/stats_ml_jun2020/master/BusinessReport-5-15-20%20(1).csv")
```

```python
df.head()
```

```python
df.columns
```

```python
#to add second and third columns, first we need to remove the $ from both columns
```

```python
#First create a new data set consisiting of only required three columns
```

```python
df_new = df[df.columns[0:3]]
```

```python
df_new
```

```python
#Define a function to get rid of $
def money_to_float(money_str):
    """ 
        Function to replace $ with space
    """
    
    return float(money_str.replace("$",""))
```

```python
pd.to_datetime(df["Date"])
```

```python
#Other method
def convert_date_format(date_str):
    """
        Function to format the date to YYYY-MM-DD
    """
    return datetime.datetime.strptime(date_str, '%m/%d/%Y').strftime('%Y-%m-%d')
```

```python
convert_date_format("4/28/2020")
```

```python
df_new1 = df_new.copy()
df_new1['Ordered Product Sales'] = df_new1['Ordered Product Sales'].apply(money_to_float)
df_new1['Ordered Product Sales - B2B'] = df_new1['Ordered Product Sales - B2B'].apply(money_to_float)
```

```python
df_new1
```

```python
# create a new column like Total_Product_sales

df_new1['Total_Prodcut_sales'] = df_new1['Ordered Product Sales'] + df_new1['Ordered Product Sales - B2B']
```

```python
df_new1
```

```python
#using the fucniton convert the date format

df_new1['New_date_format'] = df_new1['Date'].apply(convert_date_format)
```

```python
df_new1
```

```python
df_new1['Dummy_Total_Prodcut_Sales'] = 'Total_Prodcut_sales'
```

```python
df_new1
```

```python
df_final = df_new1[df_new1['Total_Prodcut_sales'] > 0]
```

```python
df_final
```

```python
df_final.columns
```

```python
df_final_new = df_final[['Dummy_Total_Prodcut_Sales','New_date_format','Total_Prodcut_sales']]
```

```python
#0r
column_order = ['Dummy_Total_Prodcut_Sales','New_date_format','Total_Prodcut_sales']
df_final = df_final[column_order]
```
